### Módelo 
$\vec{R(t)}=X(t)\vec{i}+Y(t)\vec{j}$

$X(t)=x_i+ v_xt$      

$Y(t)=y_i+v_yt-\frac{1}{2}gt^{2}$

$v_y=v_isin(\phi)$

$v_x=v_icos(\phi)$

Note que la posicion en el tiempo de caida es:

 $Y(t_c)=0$  por lo tanto 
$-y_i=t(v_y-\frac{1}{2}gt)$

donde $t=\frac{2(v_y+y_i)}{g}$


ya que $t=-y_i$ carece de sentido físico

```python
#Biblioteca matematica
import numpy as np
#Herramientas graficas
import matplotlib.pyplot as plt
```

```python
#Condiciones iniciales 
xi=float(input("Digite su posicion horizontal inicial"))
yi=float(input("Digite su posicion vertical incial"))
vi=float(input("Digite la velocidad incial"))
teta=float(input("Digite el angulo"))
n=int(input("Digite n"))
g=9.8

#Imprimimos las condiciones iniciales propuestas por el usuario
print(xi,yi,vi,teta,n)
```

```python
#Convertir a radianes
teta_rad=np.deg2rad(teta)

#Componentes de la velocidad
vx=vi*np.cos(teta_rad)
vy=vi*np.sin(teta_rad)

#Tiempo de caida 
tdc=( 2*(vy+yi) )/(g)
#Dividinos el tiempo de caida en n partes iguales 
#el cual lo usaremos para el alimento del arange, lo llamaremos incremento
incremento=tdc/n
t1=np.arange(0,tdc,incremento)

```

```python
#Funciones del modelo 
def y(t):
  y1=yi+vy*t-(1/2)*g*t**2
  return y1

def x(t):
    x1=xi+vx*t+0
    return x1
  
def vyy(t):
    vy1=vy-g*t
    return vy1

def vxx(t):
    vx1=vx-0*t
    return vx1
```

```python
#Grafica de las velocidades con respecto al tiempo
plt.figure()
plt.plot(t1,vxx(t1),"r-",linewidth=1.5, linestyle="-",label="vx")
plt.plot(t1,vyy(t1),"b-",linewidth=1.5, linestyle="-",label="vy")
plt.ylabel('vx o vy ')
plt.xlabel('Tiempo')
plt.legend(loc='upper right')
```

```python
#Grafica de la trayectoria con sus vectores velocidad 
plt.plot(x(t1),y(t1),"k-",linewidth=0.5)
#(inicio del vector x, incio del vector en y,componente x, componente y)
plt.quiver(x(t1), y(t1), vxx(t1), np.zeros(len(t1)), scale_units="xy",angles='xy', scale=11) 
plt.quiver(x(t1), y(t1), np.zeros(len(t1)), vyy(t1), scale_units="xy",angles='xy', scale=11,color='b') 
plt.quiver(x(t1), y(t1), vxx(t1), np.zeros(len(t1)), scale_units="xy",angles='xy', scale=11,color='r') 


plt.ylabel('y')
plt.xlabel('x')
plt.title("GRAFICA CHIMBITA")


plt.grid() #cuadricula  
plt.xlim(-min(x(t1)),max(x(t1))+incremento) #limite x
plt.ylim(-min(y(t1)),max(y(t1))+incremento) #limite y
```

```python
#Imprimimos la velocidad y el tiempo de caida
print("Su velocidad en y es:\n",vyy(t1))
print("Su velocidad en x es: \n",vxx(t1))
print("Su tiempo de caida es: \n",tdc)
```
